```python
# load dataset
df = pd.read_csv("Dataset_Subsystem 2.csv", header=None)
#display(df)
#IP = df[:,0:4].astype(float)
IP=df.loc[:,4:]
IP.columns = IP.iloc[0]
IP = IP[1:]
IP=IP.apply(pd.to_numeric)
display(IP)
OP=df.loc[:, 2:3]
#display(OP)
OP=pd.DataFrame(OP[2].str.cat(OP[3],sep="_"))
#display(OP)
OP = pd.DataFrame(OP.rename(columns={2: "labels"}))
OP = OP.iloc[1:]
#display(OP)
#display(OP.columns)
#display(OP.head(1).combine_first(OP.tail(1)))
#display(df.loc[:, 'camera_facing_side':'gesture'])
#df_new = OP.rename(columns={2: 'a'})
display(OP)
```

```python

num_samples = OP.shape[0]
print(OP.iloc[2].labels)
label=np.ones((num_samples,),dtype = int)

#print(num_samples)
for i in range(num_samples):
    print(i)
    print(OP.iloc[i].labels)
    if(OP.iloc[i].labels == "open_palm"):
        label[i] = 0
    if(OP.iloc[i].labels == "open_dorsal"):
        label[i] = 1
    if(OP.iloc[i].labels == "fist_palm"):
        label[i] = 2
    if(OP.iloc[i].labels == "fist_dorsal"):
        label[i] = 3
    if(OP.iloc[i].labels == "three_fingers_palm"):
        label[i] = 4
    if(OP.iloc[i].labels == "three_fingers_dorsal"):
        label[i] = 5
    
```

```python
df = pd.DataFrame({'Labels' : label})
df = df.replace({'Labels': {0: "open_palm", 1: "open_dorsal", 2: "fist_palm", 3: "fist_dorsal", 4: "three_fingers_palm", 5: "three_fingers_dorsal"}})
df.to_csv("IIS_Labels.csv",index=False)
```

```python
from sklearn.model_selection import train_test_split
 
X_train, X_test, y_train, y_test = train_test_split(IP
                                                    , df.Labels
                                                    , test_size=0.20
                                                    , shuffle=True
                                                    , random_state=41
                                                   )

y_train = np.array(y_train)
y_test = np.array(y_test)

lb = LabelEncoder()
y_train = np_utils.to_categorical(lb.fit_transform(y_train))
y_test = np_utils.to_categorical(lb.fit_transform(y_test))


print(X_train.shape)
print(y_train.shape)
```

```python
X_train = np.array(X_train)
#y_train = np.array(y_train)
X_test = np.array(X_test)
#y_test = np.array(y_test)


X_train = X_train.reshape(-1,80,1)
X_test = X_test.reshape(-1,80,1)
print(X_train.shape)
print(X_test.shape)
```

```python
model = Sequential()
model.add(Conv1D(filters=8, kernel_size=3, activation='relu', input_shape=(80,1), padding='same'))
model.add(Dropout(0.6))
model.add(MaxPooling1D(pool_size=2))
model.add(Conv1D(filters=16, kernel_size=3, activation='relu', padding='same'))
model.add(Dropout(0.6))
model.add(MaxPooling1D(pool_size=2))
model.add(Conv1D(filters=32, kernel_size=3, activation='relu', padding='same'))
model.add(Dropout(0.6))
model.add(MaxPooling1D(pool_size=2))
model.add(Flatten())
model.add(Dense(64, activation='relu'))
model.add(Dense(6, activation='softmax'))
model.compile(loss='categorical_crossentropy', optimizer='adam', metrics=['accuracy'])
```

```python
#early_stopping = EarlyStopping(monitor='val_accuracy', patience=30, verbose=1, mode='max')
early_stopping = EarlyStopping(monitor='val_loss', patience=5, verbose=1, mode='min')
mc = ModelCheckpoint('bestModel.h5', monitor='val_loss', mode='min', verbose=1,save_best_only=True)
```

```python
model_train = model.fit(X_train, y_train, batch_size=20000,epochs=10,verbose=1,validation_data=(X_test, y_test), callbacks=[early_stopping, mc])
```
